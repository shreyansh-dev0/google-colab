# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')

        for next_city in range(n):
            if mask & (1<<next_city) == 0:
                new_cost = graph[city][next_city] + visit(next_city, mask | (1<<next_city))
                if new_cost < ans:
                    ans = new_cost

        dp[city][mask] = ans
        return ans

    return visit(0, 1)
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

## Exercise 1: Print path of cities visited

```python
def tsp_with_path(graph):
    n = len(graph)
    dp = [[-1] * (1 << n) for _ in range(n)]
    parent = [[-1] * (1 << n) for _ in range(n)]

    def visit(city, mask):

        if mask == (1 << n) - 1:
            return graph[city][0]


        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')
        best_next_city = -1


        for next_city in range(n):
            if not (mask & (1 << next_city)):
                current_cost = graph[city][next_city] + visit(next_city, mask | (1 << next_city))
                if current_cost < ans:
                    ans = current_cost
                    best_next_city = next_city

        dp[city][mask] = ans
        parent[city][mask] = best_next_city
        return ans

    min_cost = visit(0, 1)


    path = [0]
    current_city = 0
    current_mask = 1
    while True:
        next_city = parent[current_city][current_mask]
        if next_city == -1 or current_mask == (1 << n) - 1:
            break
        path.append(next_city)
        current_mask |= (1 << next_city)
        current_city = next_city
    path.append(0)

    return min_cost, path



original_graph = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

min_cost, path = tsp_with_path(original_graph)
print("Minimum travelling cost:", min_cost)
print("Path of cities visited:", path)
```

## Exercise 2: Take graph input from user

```python
def get_user_graph():
    while True:
        try:
            n = int(input("Enter the number of cities: "))
            if n <= 1:
                print("Number of cities must be greater than 1. Please try again.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter an integer.")

    user_graph = []
    print(f"Enter the distance matrix ({n}x{n} rows). Separate values with spaces:")
    for i in range(n):
        while True:
            try:
                row_str = input(f"Enter row {i+1}: ")
                row = list(map(int, row_str.split()))
                if len(row) != n:
                    print(f"Row must have {n} values. Please re-enter row {i+1}.")
                    continue
                if any(dist < 0 for dist in row):
                    print("Distances cannot be negative. Please re-enter row {i+1}.")
                    continue
                user_graph.append(row)
                break
            except ValueError:
                print("Invalid input. Please enter integers separated by spaces.")


    for i in range(n):
        if user_graph[i][i] != 0:
            print(f"Warning: Distance from city {i} to itself (user_graph[{i}][{i}]) is not 0. Setting to 0.")
            user_graph[i][i] = 0

    return user_graph


user_defined_graph = get_user_graph()

print("\nUser-defined graph:")
for row in user_defined_graph:
    print(row)


if len(user_defined_graph) > 0:
    user_min_cost, user_path = tsp_with_path(user_defined_graph)
    print("\nMinimum travelling cost (user graph):", user_min_cost)
    print("Path of cities visited (user graph):", user_path)
else:
    print("No graph defined by user.")
```
