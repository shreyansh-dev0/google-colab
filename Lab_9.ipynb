# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    predecessors = {}
    visited = []


    all_nodes = set(graph.keys())
    for node_key in graph:
        for neighbor_node in graph[node_key]:
            all_nodes.add(neighbor_node)

    for node in all_nodes:
        distances[node] = float('inf')
        predecessors[node] = None

    distances[start] = 0

    while len(visited) < len(all_nodes):

        min_node = None
        min_distance = float('inf')

        for node in all_nodes:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        if min_node is None:
            break

        visited.append(min_node)


        if min_node in graph:
            for neighbor, weight in graph[min_node].items():
                new_distance = distances[min_node] + weight

                if new_distance < distances[neighbor]:
                    distances[neighbor] = new_distance
                    predecessors[neighbor] = min_node

    return distances, predecessors

graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 5, 'D': 10, 'E': 3},
    'C': {'A': 2, 'B': 5, 'D': 3},
    'D': {'B': 10, 'C': 3, 'E': 6},
    'E': {'B': 3, 'D': 6, 'F': 8},
    'F': {'E': 8}
}

# function for shortest path
def reconstruct_path(predecessors, start_node, end_node):
    path = []
    current = end_node
    while current is not None:
        path.insert(0, current)
        current = predecessors[current]
    if path and path[0] == start_node:
        return path
    return [] # Path not found


def get_user_graph():
    user_graph = {}
    print("Enter graph details. Type 'done' for nodes when finished.")
    while True:
        node = input("Enter a single node name (e.g., A) or 'done': ").upper()
        if node == 'DONE':
            break
        if node not in user_graph:
            user_graph[node] = {}
        print(f"Enter neighbors and weights for node {node}. Type 'done' for neighbors when finished.")
        while True:
            neighbor_input = input(f"Enter a single neighbor for {node} in 'Neighbor:Weight' format (e.g., B:5) or 'done': ").upper()
            if neighbor_input == 'DONE':
                break
            try:
                neighbor, weight_str = neighbor_input.split(':')
                weight = int(weight_str)
                user_graph[node][neighbor] = weight
            except ValueError:
                print("Invalid input format. Please use 'Neighbor:Weight' (e.g., B:5).")
    return user_graph
```

```python

print("\n--- Testing with Predefined Graph ---")
start_node_predefined = 'A'
distances_predefined, predecessors_predefined = dijkstra(graph, start_node_predefined)

print(f"Shortest distances from {start_node_predefined}:\n{distances_predefined}")

print("\nShortest Paths:")
for node in distances_predefined:
    if node != start_node_predefined:
        path = reconstruct_path(predecessors_predefined, start_node_predefined, node)
        if path:
            print(f"Path from {start_node_predefined} to {node}: {' -> '.join(path)} (Total distance: {distances_predefined[node]}) ")
        else:
            print(f"Path from {start_node_predefined} to {node}: No path found")


#  user input graph
print("\n--- Enter your own graph (type 'done' when finished) ---")
user_defined_graph = get_user_graph()

if user_defined_graph:
    print("\nYour graph:")
    for node, edges in user_defined_graph.items():
        print(f"{node}: {edges}")

    while True:
        start_node_user = input("Enter a source node from your graph to test (or 'done' to finish): ").upper()
        if start_node_user == 'DONE':
            break
        if start_node_user in user_defined_graph:
            distances_user, predecessors_user = dijkstra(user_defined_graph, start_node_user)

            print(f"\nShortest distances from {start_node_user}:\n{distances_user}")

            print("\nShortest Paths:")
            for node in distances_user:
                if node != start_node_user:
                    path = reconstruct_path(predecessors_user, start_node_user, node)
                    if path:
                        print(f"Path from {start_node_user} to {node}: {' -> '.join(path)} (Total distance: {distances_user[node]}) ")
                    else:
                        print(f"Path from {start_node_user} to {node}: No path found")
        else:
            print("Source node not in your graph. Please try again.")
else:
    print("No user-defined graph was entered.")
```
